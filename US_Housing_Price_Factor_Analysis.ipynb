```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
import numpy as np
```

# Importing and looking into Data

```python
df=pd.read_csv('USA_Housing.csv')
```

# Visual Representation of Data

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linecolor='white', linewidth=3)
```

# U can Observe the Various Relations 

# Now let us see the realtion one by one

# Avg. Area Income

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = df['Avg. Area Income'], y= df.Price)
plt.title('Relationship between Income and US House Prices')
```

# Area Population

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = df['Area Population'], y= df.Price)
plt.title('Relationship between Area Population and US House Prices')
```

# House Condition Avg. No of Rooms and Avg.Number of Bedrooms

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = df['Avg. Area Number of Rooms'], y= df.Price)
plt.title('Relationship between Avg. Area Number of Rooms and US House Prices')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = df['Avg. Area Number of Bedrooms'], y= df.Price)
plt.title('Relationship between Avg. Area Number of Bedrooms and US House Prices')
```

# Average Area House Age

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = df['Avg. Area House Age'], y= df.Price)
plt.title('Relationship between Avg. Area House Age and US House Prices')
```
